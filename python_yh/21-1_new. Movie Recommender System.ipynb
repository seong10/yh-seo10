## 위의 추천영화 작업을 자동화 하기 위한 파이프라인을 만드시오. 

힌트 : 반복문을 사용하여 비슷한영화에 대한 데이터프레임을 만들고, 이를 아래 빈 데이터프레임에 계속하여 추가하시오. 반복문이 끝나면, 아래 데이터프레임을 wegiht 컬럼으로 정렬하면 됩니다.

```python
similar_movies_list = pd.DataFrame()
```

```python
similar_movies_list
```

```python
myRatings
```

```python
# 1. 내가 본 영화의 이름을 가져온다.
# 내가 본 영화는 여러개 일수 있기 때문에, 반복문을 사용한다.

for i in range( myRatings.shape[0] ) :
  movie_name = myRatings['Movie Name'][i]
  recom_movies = movie_corr[movie_name].dropna().sort_values(ascending=False).to_frame()
  recom_movies.columns = ['correlation']
  recom_movies['weight'] = myRatings['Ratings'][i] * recom_movies['correlation']
  similar_movies_list = similar_movies_list.append( recom_movies )
```

```python
similar_movies_list
```

```python
# 1. weight 로 정렬한다. 이 이유는? 내가 준 별점이 반영된 컬럼이 weight 컬럼이니까.
similar_movies_list = similar_movies_list.sort_values('weight' , ascending=False)
```

```python
similar_movies_list
```

```python
# 2. 내가 본 영화는, 이 데이터프레임에서 삭제한다.
```
