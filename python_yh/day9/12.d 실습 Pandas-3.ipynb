별점 시스템을 만들려고 한다. 따라서 별점에 대한 데이터가 필요하다.

별점은 1,2,3 즉 3개로 만들것이다.

포인트가 95점 이상이면 3점, 85점 이상이면 2점, 나머지는 1점으로 할 것이다. 

리뷰데이터를 통해 각 데이터의 별점을 구하시오.

```python
df.head(3)
```

```python
get_stars(87)
# 2
```

```python
get_stars(95)
# 3

# @@ 호출부분을 먼저 써놓고
# @@ 정의부분을 해라
# @@ 방법을 익히자
```

```python
def get_stars(point) :
    if point >= 95 :
        return 3
    elif point >= 85 :
        return 2
    else :
        return 1
```

```python
df['points'].apply( get_stars )
```

```python
df['star'] = df['points'].apply( get_stars )
```

```python
df.head(3)
```
